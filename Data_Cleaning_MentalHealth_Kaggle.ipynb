```python

import numpy as np
import pandas as pd

data= pd.read_csv("C:/Users/saksh/Desktop/Personal Projects/Mental Health/survey.csv")
data.head()
#recoding the no_employees with correct order
data['no_employees']=data['no_employees'].replace(['25-Jun'], '6-25')
data['no_employees']=data['no_employees'].replace(['5-Jan'], '1-5')
```

```python
data.head(5)
data.shape
```

```python
#removing all ages below 18, there are only 6 values 
data=data[data['Age'] >= 18]
data.shape
```

```python
data['Age'].unique()
#clearly 999999 is an arbitary number. Hence we need to remove that as well
data=data[data['Age'] != 99999999999]
data.shape
```

```python
data['Gender'].unique()
```

```python
data['Gender_bin'] = data.Gender.apply(lambda x: 'Male' if x in ['M', 'Male', 'male', 'm', 'Male-ish', 'maile',
       'something kinda male?',
       'Cis Male', 'Mal', 'Male (CIS)',
       'Guy (-ish) ^_^', 'male leaning androgynous',
       'Male ', 'Man','msle',
       'queer','Mail', 'cis male', 'Malr',
       'Cis Man', 'ostensibly male, unsure what that really means'] else ('Female' if x in ['Female','female','Trans-female', 'Cis Female', 'F', 
       'Woman', 'f', 'queer/she/they',
       'Femake', 'woman', 'Female ', 
       'cis-female/femme', 'Female (trans)',
       'Female (cis)','femail'] else 'Others'))

#getting the value counts of each gender bin. This transformation is required as making dummy variables for "Gender" would have been a redundant effort
data['Gender_bin'].value_counts()
data.drop(['Gender'], axis=1)

```

```python
#getting the unique values in a dataset
pd.value_counts(data.Country)
```

```python
#calculating the number of missing values in % terms
percent_missing = data.isnull().sum() * 100 / len(data)
missing_value_data = pd.DataFrame({'column_name': data.columns,
                                 'percent_missing': percent_missing})
print(missing_value_data)
```

```python
#dropping countries and state since the data is skewed and presents a biased picture
data.drop(['state'], axis=1)
data.drop(['Country'], axis=1)
```

```python
pd.value_counts(data.self_employed)
# we should prolly replace the missing values or delete them or just recode them to a third value
#lets go with the third since there are a lot of features with "i dont know"
data = data.fillna("Unknown")
```

```python
data['self_employed'].unique()
```

```python
data['no_employees'].unique()
```

```python
#need to recode the no_employees to numerical ranked values
replace_map = {'no_employees': {'1-5': 1, '6-25': 2, '26-100': 3, '100-500': 4,
                                  '100-500': 5, '500-1000': 6, 'More than 1000': 7 }}
data.replace(replace_map, inplace=True)

print(data.head())
data['no_employees'].unique()
```

```python
#checking for the unique values in a columns
data['work_interfere'].unique()
```

```python
#replacing all "dont knows,not sure" etc with unknown to maintain uniformity in data
data.benefits.replace("Don't know","Unknown",inplace=True)
data.care_options.replace("Not sure","Unknown",inplace=True)
data.wellness_program.replace("Don't know","Unknown",inplace=True)
data.seek_help.replace("Don't know","Unknown",inplace=True)
data.anonymity.replace("Don't know","Unknown",inplace=True)
data.mental_health_consequence.replace("Maybe","Unknown",inplace=True)
data.phys_health_consequence.replace("Maybe","Unknown",inplace=True)
data.coworkers.replace("Some of them","Unknown",inplace=True)
data.supervisor.replace("Some of them","Unknown",inplace=True)
data.mental_health_interview.replace("Maybe","Unknown",inplace=True)
data.phys_health_interview.replace("Maybe","Unknown",inplace=True)
data.mental_vs_physical.replace("Don't know","Unknown",inplace=True)
```

```python
#need to recode "leaves" to numerical attributes. It could be termed as a ranked attribute
replace_map = {"leave": {"Somewhat easy": 2, "Don't know": 0, "Somewhat difficult": 3, "Very difficult": 4,
                                  "Very easy": 1}}
data.replace(replace_map, inplace=True)
```
