```python
y=df.loc[:,"Class"]
X=df.drop(["Class"],axis=1)
```

### Scale data 

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
scaled_X = scaler.transform(X)
```

```python
df_std_X = pd.DataFrame(scaled_X, columns=X.columns)
df_std_X.head()
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_std_X, 
           y, test_size=0.30, 
            random_state=11)
```

```python
X_train_SMOTE,y_train_SMOTE = SMOTE(random_state=10, k_neighbors=100).fit_resample(X_train,y_train)
```

```python
X_test_SMOTE,y_test_SMOTE = SMOTE(random_state=10, k_neighbors=100).fit_resample(X_test,y_test)
```

```python
print(len(X_test_SMOTE),len(y_test_SMOTE))
```

```python
print(len(X_train_SMOTE),len(y_train_SMOTE))
```

```python
(unique, counts) = np.unique(y_train_SMOTE, return_counts=True)
frequencies = np.asarray((unique, counts)).T
```

```python
frequencies
```

### Applying Logistic Classification 

```python
logmodel = LogisticRegression(max_iter=100)
logmodel.fit(X_train_SMOTE, y_train_SMOTE)
```

```python
Predictions = logmodel.predict(X_test_SMOTE)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_SMOTE,Predictions))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_SMOTE, Predictions))
```

### Lets go for SVM now 

```python
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear')

#Train the model using the training sets
clf.fit(X_train_SMOTE, y_train_SMOTE)

#Predict the response for test dataset
y_pred = clf.predict(X_test_SMOTE)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
